## Load data

```python
instance_data_path = (INSTANCE_DATA_FJSSPW_PATH / 'data.csv').resolve()

data = pd.read_csv(instance_data_path)

header = data.columns
```

## Filter function

```python
def filter(bounds : Dict[str, Tuple[float|int, float|int]]) -> pd.DataFrame:
    filtered_data = data.copy()
    for bound in bounds:
        lb = bounds[bound][0]
        ub = bounds[bound][1]
        filtered_data = filtered_data[
            lb <= filtered_data[bound] <= ub
        ]
    return filtered_data
```

## The hardest (large, varied, very flexible)

## Save to file

```python
instances = {
    "small": small_instances,
    "medium": medium_instances,
    "large": large_instances,
    "dense": dense_instances,
    "varied": varied_instances,
    "hardest": hardest_instances,
}

for name, instance in instances.items():
    instance['type'] = name

selected_instances = pd.concat(instances.values())
selected_instances.to_csv(Path("instances/SelectedInstanceData/selected_instances_data.csv"), index=False)
```
