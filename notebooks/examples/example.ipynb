```python
from util.greedy_solver import GreedyFJSSPSolver, GreedyFJSSPWSolver
import numpy as np
from pprint import pprint
```

```python
from util.benchmark_parser import WorkerBenchmarkParser
parser = WorkerBenchmarkParser()
encoding = parser.parse_benchmark(r'instances/Example_Instances_FJSSP-WF/Fattahi1.fjs')
```

```python
solver = GreedyFJSSPWSolver(encoding.durations(), encoding.job_sequence())
sequence, machines, workers = solver.solve()
```

```python
print("Durations:", len(encoding.durations()))
print("Jobs:", len(encoding.job_sequence()))
print("Unique jobs:", len(np.unique(encoding.job_sequence())))
print("Durations: ", end="")
pprint(encoding.durations())
print("Job sequence:", list(encoding.job_sequence()))
print("Unique durations:", sorted(np.unique(encoding.durations())))
print("Unique job sequence values:", sorted(np.unique(encoding.job_sequence())))
```

```python
print("Sequence:", len(sequence), "Machines:", len(machines), "Workers:", len(workers))
print("Sequence:", sequence)
print("Machines:", machines)
print("Workers:", workers)
```

```python
import util.evaluation as evaluation
start_times, machines, workers = evaluation.translate(sequence, machines, workers, encoding.durations()) # necessary because the solver does not provide the start times of the schedule
c = evaluation.makespan(start_times, machines, workers, encoding.durations())
```
